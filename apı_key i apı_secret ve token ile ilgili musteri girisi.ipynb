```python
import requests

api_key = "your_api_key"
api_secret = "your_api_secret"
token = "your_token"

def authenticate(api_key, api_secret):
    url = "https://api.trendyol.com/authenticate"

    headers = {
        "Content-Type": "application/json"
    }

    payload = {
        "api_key": api_key,
        "api_secret": api_secret
    }

    try:
        response = requests.post(url, headers=headers, json=payload)
        response.raise_for_status()  

        token = response.json().get("access_token")
        return token
    except requests.exceptions.RequestException as e:
        print("Kimlik doğrulama işlemi sırasında bir hata oluştu:")
        print(e)
        return None

def make_authenticated_request(url, token):
    headers = {
        "Content-Type": "application/json",
        "Authorization": f"Bearer {token}"
    }

    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # HTTP hatalarını kontrol et

        data = response.json()
        return data
    except requests.exceptions.RequestException as e:
        print("İstek sırasında bir hata oluştu:")
        print(e)
        return None

# Kimlik doğrulama
token = authenticate(api_key, api_secret)

if token:
    endpoint = "https://api.trendyol.com/some_endpoint"
    response_data = make_authenticated_request(endpoint, token)

    if response_data:
        print("API yanıtı:")
        print(response_data)
```
