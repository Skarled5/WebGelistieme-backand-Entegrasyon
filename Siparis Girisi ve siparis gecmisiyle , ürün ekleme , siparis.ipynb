```python
import requests
import random
import sqlite3

API_URL = "https://api.example.com/orders"  # Gerçek API URL'sini buraya yerleştirin

# Veritabanı bağlantısı oluştur
conn = sqlite3.connect("e_ticaret.db")
cursor = conn.cursor()

# Tabloları oluştur
cursor.execute("""
    CREATE TABLE IF NOT EXISTS urunler (
        urun_id INTEGER PRIMARY KEY AUTOINCREMENT,
        urun_adi TEXT NOT NULL,
        fiyat REAL,
        stok INTEGER
    )
""")

cursor.execute("""
    CREATE TABLE IF NOT EXISTS siparisler (
        siparis_id INTEGER PRIMARY KEY AUTOINCREMENT,
        ad TEXT,
        soyad TEXT,
        email TEXT,
        telefon TEXT,
        siparis TEXT
    )
""")
conn.commit()

def siparis_gecmisini_oku():
    cursor.execute("SELECT * FROM siparisler")
    siparisler = cursor.fetchall()
    return siparisler

def siparis_girisi():
    siparisler = []
    while True:
        siparis = input("Siparişinizi girin (Çıkış için 'q' tuşuna basın): ")
        if siparis.lower() == "q":
            break
        siparisler.append(siparis)
    return siparisler

def kullanici_bilgilerini_al():
    ad = input("Adınızı girin: ")
    soyad = input("Soyadınızı girin: ")
    email = input("E-posta adresinizi girin: ")
    telefon = input("Telefon numaranızı girin: ")

    kullanici_bilgileri = {
        "ad": ad,
        "soyad": soyad,
        "email": email,
        "telefon": telefon
    }

    return kullanici_bilgileri

def siparis_gecmisini_guncelle(siparisler, kullanici_bilgileri):
    ad = kullanici_bilgileri['ad']
    soyad = kullanici_bilgileri['soyad']
    email = kullanici_bilgileri['email']
    telefon = kullanici_bilgileri['telefon']
    
    for siparis in siparisler:
        cursor.execute("INSERT INTO siparisler (ad, soyad, email, telefon, siparis) VALUES (?, ?, ?, ?, ?)",
                       (ad, soyad, email, telefon, siparis))
    def siparis_apiye_gonder(siparisler):
        for siparis in siparisler:
            # API'ye gerçek bir istek göndermek yerine rastgele bir yanıt oluşturuyoruz
            # Gerçek bir API kullanıyorsanız, uygun şekilde API isteklerini düzenleyin
            payload = {
                "order_id": random.randint(1000, 9999),
                "product_name": siparis,
                "status": "completed"
            }
        response = requests.post(API_URL, json=payload)
        if response.status_code == 200:
            print(f"{siparis} başarıyla gönderildi!")
        else:
            print(f"{siparis} gönderilirken bir hata oluştu!")

def urunleri_listele():
    cursor.execute("SELECT * FROM urunler")
    urunler = cursor.fetchall()
    if not urunler:
        print("Stokta mevcut ürün bulunmamaktadır.")
    else:
        print("Stoktaki Ürünler:")
        for urun in urunler:
            print(f"Ürün ID: {urun[0]}")
            print(f"Ürün Adı: {urun[1]}")
            print(f"Fiyatı: {urun[2]} TL")
            print(f"Stok Durumu: {urun[3]} adet")
            print("")

def urun_ekle():
    urun_adi = input("Ürün adını girin: ")
    fiyat = float(input("Ürün fiyatını girin: "))
    stok = int(input("Stok miktarını girin: "))

    cursor.execute("INSERT INTO urunler (urun_adi, fiyat, stok) VALUES (?, ?, ?)",
                   (urun_adi, fiyat, stok))
    conn.commit()
    print("Ürün başarıyla eklendi.")

def siparis_gecmisi():
    cursor.execute("SELECT * FROM siparisler")
    siparisler = cursor.fetchall()
    if not siparisler:
        print("Sipariş geçmişi bulunmamaktadır.")
    else:
        print("Sipariş Geçmişi:")
        for siparis in siparisler:
            print(f"Sipariş ID: {siparis[0]}")
            print(f"Ad: {siparis[1]}")
            print(f"Soyad: {siparis[2]}")
            print(f"E-posta: {siparis[3]}")
            print(f"Telefon: {siparis[4]}")
            print(f"Sipariş: {siparis[5]}")
            print("")

def cikis():
    print("Çıkış yapılıyor...")
    conn.close()
    exit()

# Kullanıcı bilgilerini al
kullanici_bilgileri = kullanici_bilgilerini_al()

while True:
    print("\n--- E-Ticaret Sistemi ---")
    print("1. Ürünleri Listele")
    print("2. Ürün Ekle")
    print("3. Sipariş Gir")
    print("4. Sipariş Geçmişini Gör")
    print("5. Çıkış")

    secim = input("Bir seçenek girin: ")

    if secim == "1":
        urunleri_listele()
    elif secim == "2":
        urun_ekle()
    elif secim == "3":
        siparisler = siparis_girisi()
        siparis_gecmisini_guncelle(siparisler, kullanici_bilgileri)
        siparis_apiye_gonder(siparisler)
    elif secim == "4":
        siparis_gecmisi = siparis_gecmisini_oku()
        print("\n--- Sipariş Geçmişi ---")
        for siparis in siparis_gecmisi:
            print(siparis)
    elif secim == "5":
        print("Çıkış yapılıyor...")
        break
    else:
        print("Geçersiz seçenek! Lütfen tekrar deneyin")
```
