```python
from flask import Flask, request
import requests
import json

app = Flask(__name__)

class TrendyolEntegrasyonu:
    def __init__(self, api_key):
        self.api_key = api_key
    
    def stok_sorgula(self, magaza_kimlik):
        url = f"https://api.trendyol.com/sapigw/suppliers/{magaza_kimlik}/v2/products/stock"

        headers = {
            "Content-Type": "application/json",
            "Authorization": f"Basic {self.api_key}"
        }

        response = requests.get(url, headers=headers)

        if response.status_code == 200:
            stok_bilgisi = response.json()
            return stok_bilgisi
        else:
            print(f"Stok sorgulanamadı: Mağaza Kimlik - {magaza_kimlik}")
            print(f"HTTP Hata Kodu: {response.status_code}")
            print(f"API Yanıtı: {response.text}")
            return None
    
    def iade_taleplerini_al(self, magaza_kimlik):
        url = f"https://api.trendyol.com/sapigw/suppliers/{magaza_kimlik}/v2/returns"

        headers = {
            "Content-Type": "application/json",
            "Authorization": f"Basic {self.api_key}"
        }

        response = requests.get(url, headers=headers)

        if response.status_code == 200:
            iade_talepleri = response.json()
            return iade_talepleri
        else:
            print(f"Iade talepleri alınamadı: Mağaza Kimlik - {magaza_kimlik}")
            print(f"HTTP Hata Kodu: {response.status_code}")
            print(f"API Yanıtı: {response.text}")
            return None
    
    def iade_talebini_onayla(self, magaza_kimlik, iade_id):
        url = f"https://api.trendyol.com/sapigw/suppliers/{magaza_kimlik}/v2/returns/{iade_id}/approve"

        headers = {
            "Content-Type": "application/json",
            "Authorization": f"Basic {self.api_key}"
        }

        response = requests.put(url, headers=headers)

        if response.status_code == 200:
            print(f"Iade talebi onaylandı: İade ID - {iade_id}")
        else:
            print(f"Iade talebi onaylanamadı: İade ID - {iade_id}")
            print(f"HTTP Hata Kodu: {response.status_code}")
            print(f"API Yanıtı: {response.text}")
    
    def iade_talebini_reddet(self, magaza_kimlik, iade_id):
        url = f"https://api.trendyol.com/sapigw/suppliers/{magaza_kimlik}/v2/returns/{iade_id}/reject"

        headers = {
            "Content-Type": "application/json",
            "Authorization": f"Basic {self.api_key}"
        }

        response = requests.put(url, headers=headers)

        if response.status_code == 200:
            print(f"Iade talebi reddedildi: İade ID - {iade_id}")
            print(f"Iade talebi reddedilemedi: İade ID - {iade_id}")
            print(f"HTTP Hata Kodu: {response.status_code}")
            print(f"API Yanıtı: {response.text}")
    
    def fatura_olustur(self, magaza_kimlik, fatura_bilgileri):
        url = f"https://api.trendyol.com/sapigw/suppliers/{magaza_kimlik}/v2/invoices"

        headers = {
            "Content-Type": "application/json",
            "Authorization": f"Basic {self.api_key}"
        }

        response = requests.post(url, headers=headers, json=fatura_bilgileri)

        if response.status_code == 201:
            fatura_id = response.json()['id']
            print(f"Fatura oluşturuldu: Fatura ID - {fatura_id}")
        else:
            print(f"Fatura oluşturulamadı")
            print(f"HTTP Hata Kodu: {response.status_code}")
            print(f"API Yanıtı: {response.text}")
    
    def fiyat_guncelle(self, magaza_kimlik, urun_kimlik, yeni_fiyat):
        url = f"https://api.trendyol.com/sapigw/suppliers/{magaza_kimlik}/v2/products/{urun_kimlik}/prices"

        headers = {
            "Content-Type": "application/json",
            "Authorization": f"Basic {self.api_key}"
        }

        payload = {
            "price": yeni_fiyat
        }

        response = requests.put(url, headers=headers, json=payload)

        if response.status_code == 200:
            print(f"Fiyat güncellendi: Ürün Kimlik - {urun_kimlik}")
        else:
            print(f"Fiyat güncellenemedi: Ürün Kimlik - {urun_kimlik}")
            print(f"HTTP Hata Kodu: {response.status_code}")
            print(f"API Yanıtı: {response.text}")
    
    def urun_ekle(self, magaza_kimlik, urun_bilgileri):
        url = f"https://api.trendyol.com/sapigw/suppliers/{magaza_kimlik}/v2/products"

        headers = {
            "Content-Type": "application/json",
            "Authorization": f"Basic {self.api_key}"
        }

        response = requests.post(url, headers=headers, json=urun_bilgileri)

        if response.status_code == 201:
            urun_id = response.json()['id']
            print(f"Ürün eklendi: Ürün ID - {urun_id}")
        else:
            print(f"Ürün eklenemedi")
            print(f"HTTP Hata Kodu: {response.status_code}")
            print(f"API Yanıtı: {response.text}")
    
    def urun_sil(self, magaza_kimlik, urun_kimlik):
        url = f"https://api.trendyol.com/sapigw/suppliers/{magaza_kimlik}/v2/products/{urun_kimlik}"

        headers = {
            "Content-Type": "application/json",
            "Authorization": f"Basic {self.api_key}"
        }

        response = requests.delete(url, headers=headers)
```
